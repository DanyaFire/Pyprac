**Задача 1**

Учительница Анна Алгоритмовна решила попробовать создать свой калькулятор,
чтобы ускорить проверку контрольных по математике на тему «деление». Она решила, что будет достаточно реализовать функционал для деления двух действительных
чисел. Когда Анна писала код, она вспомнила, что нужно учесть запрет на деление
на 0.

Помогите Анне написать функцию, которая принимает на вход строку, содержащую арифметическое выражение из двух чисел и одного оператора деления через
пробел между ними, и возвращает одно число – результат вычисления выражения.
Если при этом выдаётся исключение ZeroDivisionError, тогда функция возвращает слово ERROR.

```python
def divide(s): #Строка 'a / b'
    L1 = s.split()
    if L1[1] != '/':
        return None
    try:
        res = float(L1[0]) / float(L1[2])
        return res
    except ZeroDivisionError:
        print('На 0 делить нельзя!')
        return 'ERROR'

print(divide('555 / 5'))
print(divide('23 / 0.5'))
print(divide('777 / 0'))
```

**Задача 2**

Пётр Паролевич в очередной раз забыл свой пароль. К счастью, он нашёл листочек, на котором записывает все свои пароли. Каждый пароль может состоять из
латинских строчных букв и десятичных цифр.

Петр Паролевич помнит, что его пароль – шестнадцатеричное число. По рассеянности Пётр Паролевич начал вводить все пароли подряд. Помогите Петру разобраться, какие строки могут являться его паролями.
Напишите функцию, которая принимает в качестве аргументов сколько угодно
строк, и возвращает список со всеми значениями, которые могут быть паролями
Петра Паролевича.
Примечание. Для проверки пароля можно использовать int(password, 16).
Если число не подходит, тогда нужно обработать исключение ValueError.


```python
def check_pass(*passwords):
    can_be_pass = []
    for pw in passwords:
        try:
            password = int(pw, 16)
            can_be_pass.append(pw)
        except ValueError:
            print(f'Что-то не то: {pw}')
    return can_be_pass

check_pass('A', "FF", "0x10", 'r')
```

Олимпиада «Горные воробьи» выкладывает такой же словарь, но под ключами

хочет заодно проверить, опубликован ли её балл за решение 5-ой задачи.

Напишите функцию, которая проверяет, есть ли ученик в списке победителей для
каждой из олимпиад, и выведите результат в следующем формате:

# последнее значение в каждой строке выводится при наличии в списке
[название_первой_олимпиады] статус балл_за_олимпиаду
--------------------
[название_второй_олимпиады] статус балл_за_пятую_задачу
--------------------

Если ученик есть в списке победителей, то он получает статус «победитель»,
иначе – статус «призёр», поскольку в этих олимпиадах «вы все уже победители».
После того как Ольга увидела свои баллы, она также решила проверить результаты сестры – Олеси Олимпиадниковой. В качестве ответа запишите результат работы
функции для Ольги и Олеси.

Примечание. В этом задании подразумевается использование конструкции
finally и проверка на исключения IndexError и KeyError

```python
def check_olymp(name):
    print(name)
    try:
        points = olympiad1['winners'][name]
        print(f"[{olympiad1['name']}] победитель {points}")
    except KeyError:
        print(f"[{olympiad1['name']}] призёр")
    finally:
        print('--------------------')
    try:
        point_for_5 = olympiad2['winners'][name][4]
        print(f"[{olympiad2['name']}] победитель {point_for_5}")
    except IndexError:
        print(f"[{olympiad2['name']}] призёр")
    finally:
        print('--------------------')

check_olymp('Ольга Олимпиадникова')
check_olymp('Олеся Олимпиадникова')
```

**Задача 5**

Реализуйте программу, которая соответствует функционалу известного анекдота.
У этого задания нет единственного способа решения. Вы можете сами выбрать, какие
исключения возвращать и что делать в каждом из случаев. Можно использовать
комментарии под строками анекдота или самостоятельно придумать чтонибудь уникальное.
Что рекомендуется использовать:
1. инструкцию raise;
2. инструкцию assert, возвращая правильные сообщения об ошибке;
3. классы типа Exception для обозначения уникальных ошибок;
4. инструкцию finally для вывода сообщения о результате действия.
Анекдот.
Заходит однажды тестировщик в бар.
Заказывает:
кружку лимонада,
2 кружки лимонада,
0 кружек лимонада,
999999999 кружек лимонада,
ящерицу в стакане,
–1 кружку лимонада,
qwerty кружек лимонада.
Первый реальный клиент заходит в бар и спрашивает, где туалет. Бар вспыхивает
пламенем.

```python
class LizardInCup(Exception):
    pass

class BarBurntDown(Exception):
    pass

def bar(x):
    try:
        if x == 'ящерица в стакане':
            raise LizardInCup
        if x == 'где туалет?':
            raise BarBurntDown
        if int(x) <= 0 or int(x) >= 100:
            raise ValueError
        assert int(x) < 2, 'У бара сегодня кризис: не больше одной кружки в одни руки'
    except AssertionError:
        print('У бара сегодня кризис: не больше одной кружки в одни руки')
    except ValueError:
        print('Вы не можете заказать столько кружек лимонада. Держите одну!')
    except LizardInCup:
        print('Извините, но у нас закончились ящерицы. Приходите завтра!')
    except BarBurntDown:
        print("Поздравляем! Бар сгорел. Вы вышли из симуляции.")
    finally:
        print('Приходите ещё!')

bar(input('Вы заходите в бар. Сколько кружек лимонада заказываете?\n').lower())
```
